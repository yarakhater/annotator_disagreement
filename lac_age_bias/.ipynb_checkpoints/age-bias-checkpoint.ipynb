```python
import pandas as pd
import numpy as np
from scipy.special import logsumexp
from sklearn.model_selection import GroupShuffleSplit 
from models import utils
from models import mace, latent_annotator_clustering
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import homogeneity_score, completeness_score, v_measure_score
```

```python
train_df = pd.read_csv('./data/train_older_adult_annotations.csv',delimiter=',', encoding='latin-1')
test_df = pd.read_csv('./data/test_annotations.csv', delimiter=',')
anxiety_score_df = pd.read_csv('./data/respondent_anxiety_table.csv')
demographics_df = pd.read_csv('./data/demographics_responses.csv', delimiter=',')

df = pd.concat([test_df, train_df])

anxiety_score_df = anxiety_score_df[anxiety_score_df['respondent_id'].isin(df['respondent_id'])]
demographics_df = demographics_df[demographics_df['respondent_id'].isin(df['respondent_id'])]

sentiment_labels = ['Very negative','Somewhat negative','Neutral','Somewhat positive','Very positive']
annotator_ids = df['respondent_id'].unique().tolist()

id2label = {index: row for (index, row) in enumerate(sentiment_labels)} 
label2id = {row: index for (index, row) in enumerate(sentiment_labels)}

id2annotator = {index: row for (index, row) in enumerate(annotator_ids)}
annotator2id = {row: index for (index, row) in enumerate(annotator_ids)}

df["annotation"] = df["annotation"].map(label2id)
df["respondent_id"] = df["respondent_id"].map(annotator2id)
anxiety_score_df["respondent_id"] = anxiety_score_df["respondent_id"].map(annotator2id)
demographics_df["respondent_id"] = demographics_df["respondent_id"].map(annotator2id)




splitter = GroupShuffleSplit(test_size=0.3, n_splits=2, random_state = 0)
split = splitter.split(df, groups = df['unit_id'])
train_inds, test_inds = next(split)


train_df = df.iloc[train_inds]
test_df = df.iloc[test_inds]

train_df = train_df.sample(frac=1)
test_df = test_df.sample(frac=1)

#keep rows in train df where is_age_related is "Yes"
train_df = train_df[train_df['is_age_related']=='Yes']

train_df = train_df[['unit_id','respondent_id','annotation']]
test_df = test_df[['unit_id','respondent_id','annotation']]

```

```python
data = list()

# Iterate over the rows of the DataFrame
for index, row in pivot_df.iterrows():
    data.append(list())
    for a in row.index:
    # Extract the annotator and annotation values from the row
        annotator = a
        annotation = int(row[a])
        # print(annotator, annotation)
        if(annotation  != -1):
        # Append the (annotator, annotation) pair to the data list
            data[-1].append((annotator, annotation))
```
